```python
import random
options = ["rock","paper","scissor"]
options
```

```python
agent_1 = random.choice(options)
agent_2 = random.choice(options)
print(agent_1)
print(agent_2)
```

```python
if agent_1 == agent_2:
    print("It's draw!")
elif agent_1 == 'rock' and agent_2 == "scissor":
    print("agent_1 won!")
elif agent_1 == "scissor" and agent_2 == "paper":
     print("agent 1 won!")
elif agent_1 == "paper" and agent_2 == "rock":
     print("agent 1 won")
else:
     print("agent_2 won")
```
